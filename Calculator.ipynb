```python
from tkinter import*

def btnClick(numbers):
    global operator
    operator=operator + str(numbers)
    text_Input.set(operator)
    
def btnClearDisplay():
    global operator
    operator=""
    text_Input.set("")
     
def btnEualsInput():
    global operator
    sumup=str(eval(operator))
    text_Input.set(sumup)
    operator=""
    
cal=Tk()
cal.title("Calculator")
operator=""
text_Input=StringVar()

textDisplay=Entry(cal,font=("chesterish",22,"bold"),textvariable=text_Input,bd=40,
                  insertwidth=5,bg="white",justify="right").grid(columnspan=5)

btn7=Button(cal,padx=16,pady=16,bd=8,fg="black",font=("chesterish",22,"bold"),
            text="7",bg="blue",command=lambda:btnClick(7)).grid(row=1,column=0)

btn8=Button(cal,padx=16,pady=16,bd=8,fg="black",font=("chesterish",22,"bold"),
            text="8",bg="white",command=lambda:btnClick(8)).grid(row=1,column=1)

btn9=Button(cal,padx=16,pady=16,bd=8,fg="black",font=("chesterish",22,"bold"),
            text="9",bg="yellow",command=lambda:btnClick(9)).grid(row=1,column=2)

Division=Button(cal,padx=16,pady=16,bd=8,fg="black",font=("chesterish",22,"bold"),
                text="/",bg="orange",command=lambda:btnClick("/")).grid(row=1,column=3)

btn4=Button(cal,padx=16,pady=16,bd=8,fg="black",font=("chesterish",22,"bold"),
            text="4",bg="yellow",command=lambda:btnClick(4)).grid(row=2,column=0)

btn5=Button(cal,padx=16,pady=16,bd=8,fg="black",font=("chesterish",22,"bold"),
            text="5",bg="blue",command=lambda:btnClick(5)).grid(row=2,column=1)

btn6=Button(cal,padx=16,pady=16,bd=8,fg="black",font=("chesterish",22,"bold"),
            text="6",bg="pink",command=lambda:btnClick(6)).grid(row=2,column=2)

Multiplication=Button(cal,padx=16,pady=16,bd=8,fg="black",font=("chesterish",22,"bold"),
                      text="X",bg="white",command=lambda:btnClick("X")).grid(row=2,column=3)

btn1=Button(cal,padx=16,pady=16,bd=8,fg="black",font=("chesterish",22,"bold"),
            text="1",bg="yellow",command=lambda:btnClick(1)).grid(row=3,column=0)

btn2=Button(cal,padx=16,pady=16,bd=8,fg="black",font=("chesterish",22,"bold"),
            text="2",bg="orange",command=lambda:btnClick(2)).grid(row=3,column=1)

btn3=Button(cal,padx=16,pady=16,bd=8,fg="black",font=("chesterish",22,"bold"),
            text="3",bg="pink",command=lambda:btnClick(3)).grid(row=3,column=2)

Addition=Button(cal,padx=16,pady=16,bd=8,fg="black",font=("chesterish",22,"bold"),
                text="+",bg="white",command=lambda:btnClick("+")).grid(row=3,column=3)

btn0=Button(cal,padx=16,pady=16,bd=8,fg="black",font=("chesterish",22,"bold"),
            text="0",bg="blue",command=lambda:btnClick(0)).grid(row=4,column=0)

btnClear=Button(cal,padx=16,pady=16,bd=8,fg="black",font=("chesterish",22,"bold"),
                text="clear",bg="red",command=btnClearDisplay).grid(row=4,column=1)

btnequalsto=Button(cal,padx=16,pady=16,bd=8,fg="black",font=("chesterish",22,"bold"),
                   text="=",bg="orange",command=btnEualsInput).grid(row=4,column=2)

Substraction=Button(cal,padx=16,pady=16,bd=8,fg="black",font=("chesterish",22,"bold"),
                    text="-",bg="white",command=lambda:btnClick("-")).grid(row=4,column=3)
cal.mainloop()
```
